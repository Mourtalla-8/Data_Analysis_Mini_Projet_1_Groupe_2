# Rapport Final du Mini Projet de Data Analysis

Ce rapport présente une analyse complète des données salariales contenues dans le dataset Salary_Data. L'objectif est de fournir des insights pertinents en utilisant des techniques de traitement des données, des mesures de tendance centrale et de dispersion, et des visualisations adaptées.

## Livrable 1. Importation des Bibliothèques Nécessaires

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Livrable 2. DataFrame contenant les données sur le salaire

### 1. Chargement du Jeu de Données

```python
# Chargement du dataset
dataset = pd.read_csv('Dataset/Salary_Data.csv')

# Affichage des cinq premières lignes pour vérification
dataset.head()
```

### 2. Vérification des Informations de Base

```python
# Vérification de la forme du dataset
print("Forme du dataset :", dataset.shape)
print()

# Vérification des types de données et des valeurs nulles
dataset.info()
```

### 3. Gestion des Valeurs Manquantes

```python
# Suppression des doublons
dataset.drop_duplicates(inplace=True)

# Vérification des valeurs manquantes par colonne
print("Valeurs manquantes par colonne :")
print(dataset.isnull().sum())

# Remplacement des valeurs manquantes par le mode pour les colonnes catégorielles
cat_columns = dataset.select_dtypes(include=['object']).columns
for col in cat_columns:
    dataset[col].fillna(dataset[col].mode()[0], inplace=True)

# Remplacement des valeurs manquantes par la médiane pour les colonnes numériques
num_columns = dataset.select_dtypes(include=['float64']).columns
for col in num_columns:
    dataset[col].fillna(dataset[col].median(), inplace=True)

# Vérification des valeurs manquantes après traitement
print("\nValeurs manquantes après traitement :")
print(dataset.isnull().sum())
```

### 4. Conversion des Types de Données

Certaines colonnes de notre dataset contiennent des valeurs répétitives et peuvent être converties en type `category` pour réduire la mémoire utilisée et améliorer les performances. Nous avons identifié les colonnes `Gender`, `Education Level` et `Job Title` pour cette conversion.

```python
# Avant de procéder à la conversion, nous allons vérifier les valeurs répétitives dans ces colonnes.

# Vérification des valeurs répétitives avant conversion
print("Valeurs répétitives avant conversion :")
for col in ['Gender', 'Education Level', 'Job Title']:
    print(f"{col} : {dataset[col].nunique()} valeurs uniques")
```

```python
# Conversion des colonnes en type catégorie
cols_to_category = ['Gender', 'Education Level', 'Job Title']
dataset[cols_to_category] = dataset[cols_to_category].astype('category')

# Vérification des types de données après conversion
print("Résumé des informations après conversion :")
dataset.info()
```

### 5. Gestion des Valeurs Aberrantes

##### Fonction de Gestion des Valeurs Aberrantes

```python
def handle_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identifier les valeurs aberrantes
    aberrant_values = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"\nNombre de valeurs aberrantes dans {column} avant traitement : {aberrant_values.shape[0]}")
    
    # Remplacement des valeurs aberrantes par la médiane
    median_value = df[column].median()
    df.loc[(df[column] < lower_bound) | (df[column] > upper_bound), column] = median_value
    
    # Vérification des valeurs aberrantes après traitement
    aberrant_values_after = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Nombre de valeurs aberrantes dans {column} après traitement : {aberrant_values_after.shape[0]}")
    
    return df

# Application de la fonction aux colonnes concernées
dataset = handle_outliers(dataset, 'Salary')
dataset = handle_outliers(dataset, 'Age')
dataset = handle_outliers(dataset, 'Years of Experience')
```

### 6. Statistiques Descriptives

```python
# Résumé statistique des variables numériques
print("Résumé statistique des variables numériques :")
dataset.describe()
```

```python
# Résumé statistique des variables catégorielles
print("\nRésumé statistique des variables catégorielles :")
dataset.describe(include='category')
```

## Livrable 3. Calcul des Mesures de Tendance Centrale

```python
# Salaire
mean_salary = dataset['Salary'].mean()
median_salary = dataset['Salary'].median()
mode_salary = dataset['Salary'].mode()[0]

# Âge
mean_age = dataset['Age'].mean()
median_age = dataset['Age'].median()
mode_age = dataset['Age'].mode()[0]

# Années d'Expérience
mean_experience = dataset['Years of Experience'].mean()
median_experience = dataset['Years of Experience'].median()
mode_experience = dataset['Years of Experience'].mode()[0]


print("Mesures de Tendance Centrale :")
print(f"\nSalaire - Moyenne : {mean_salary}, Médiane : {median_salary}, Mode : {mode_salary}")
print(f"\nÂge - Moyenne : {mean_age}, Médiane : {median_age}, Mode : {mode_age}")
print(f"\nExpérience - Moyenne : {mean_experience}, Médiane : {median_experience}, Mode : {mode_experience}")
```

## Livrable 4. Calcul des Mesures de Dispersion

```python
# Salaire
std_salary = dataset['Salary'].std()
var_salary = dataset['Salary'].var()
min_salary = dataset['Salary'].min()
max_salary = dataset['Salary'].max()
range_salary = max_salary - min_salary
iqr_salary = dataset['Salary'].quantile(0.75) - dataset['Salary'].quantile(0.25)

# Âge
std_age = dataset['Age'].std()
var_age = dataset['Age'].var()
min_age = dataset['Age'].min()
max_age = dataset['Age'].max()
range_age = max_age - min_age
iqr_age = dataset['Age'].quantile(0.75) - dataset['Age'].quantile(0.25)

# Années d'Expérience
std_experience = dataset['Years of Experience'].std()
var_experience = dataset['Years of Experience'].var()
min_experience = dataset['Years of Experience'].min()
max_experience = dataset['Years of Experience'].max()
range_experience = max_experience - min_experience
iqr_experience = dataset['Years of Experience'].quantile(0.75) - dataset['Years of Experience'].quantile(0.25)


print("Mesures de Dispersion :")
print(f"\nSalaire - Écart-Type : {std_salary}, Variance : {var_salary}, Minimum : {min_salary}, Maximum : {max_salary}, Amplitude : {range_salary}, IQR : {iqr_salary}")
print(f"\nÂge - Écart-Type : {std_age}, Variance : {var_age}, Minimum : {min_age}, Maximum : {max_age}, Amplitude : {range_age}, IQR : {iqr_age}")
print(f"\nExpérience - Écart-Type : {std_experience}, Variance : {var_experience}, Minimum : {min_experience}, Maximum : {max_experience}, Amplitude : {range_experience}, IQR : {iqr_experience}")
```

#### Répartition par Titre de Poste

Pour éviter que les titres de poste ne soient trop nombreux et mélangés, nous pouvons afficher les 10 titres les plus fréquents.
